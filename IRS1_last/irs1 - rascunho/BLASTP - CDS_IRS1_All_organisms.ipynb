```python
cds = SeqIO.read("CDS_IRS1.fasta","fasta") 
result_cds_allOrg = NCBIWWW.qblast("blastp", "swissprot", cds.format("fasta"))
save_file_cds = open("CDS_IRS1_allOrganisms.xml", "w")
save_file_cds.write(result_cds_allOrg.read()) 
save_file_cds.close() 
result_cds_allOrg.close()
```

```python
cds_xml = open("CDS_IRS1_allOrganisms.xml")
cds_all= NCBIXML.parse(cds_xml)
for parametros in cds_all:
    for info in parametros.alignments:
        print("ID accession: ", info.accession) 
        print('description: ', info.hit_def)
        for hsp in info.hsps:
            print('identity: ',hsp.identities)
            print('E-value: ', hsp.expect)
            print('alignment length: ', hsp.align_length)
            print()
```

```python
E_VALUE_THRESH = 0.001
results_Blast= open('CDS_IRS1_allOrganisms.xml')
blast_records = NCBIXML.read(results_Blast)
for alignment in  blast_records.alignments:
    #print(alignment.accession)
    for hsp in alignment.hsps:
        if hsp.expect < E_VALUE_THRESH:
            print('****Alignment****')
            print('sequence: ', alignment.title)
            print('lenght:', alignment.length)
            print(hsp.query[0:75] + '...')
            print(hsp.match[0:75] + '...')
            print(hsp.sbjct[0:75] + '...')
            print()
```

```python
import re

results_Blast= open('CDS_IRS1_allOrganisms.xml')
blast_records = NCBIXML.read(results_Blast)
E_VALUE_THRESH = 0.001
list_filtered_alignments,list_species=[],[]
for alignment in  blast_records.alignments:
    for hsp in alignment.hsps:
        if hsp.expect < E_VALUE_THRESH:
            list_filtered_alignments.append(alignment.accession)
            title_organism=re.search(r'\[.+\s.+\]', alignment.title)
            if title_organism:
                m = re.match(r'\[.+\s.+\]', title_organism[0] )
                specie = m.group(0)
                #print(specie)
                list_species.append(specie)
for x in sorted(set(list_species)):
    print("nº de vezes: {} que surgiu a espécie: {})".format(list_species.count(x),x))

print(list_filtered_alignments)
```

```python
with open('CDS_protein_result_blast.txt', 'w') as f:
    for line in list_filtered_alignments:
        f.write(f"{line}\n")
```
