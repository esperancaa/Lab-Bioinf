```python
from Bio.Blast import NCBIXML 
from Bio.Blast import NCBIWWW 
from Bio import SeqIO
```

```python
record = SeqIO.read(open("3729bp_sequence_DNA_IRS1.gb"), format="gb") 
print (len(record.seq))
```

```python
Blast = NCBIWWW.qblast("blastn", "nt", record.seq)
```

```python
with open('blast_DNA_IRS1_file.xml', "w") as out_handle:
        out_handle.write(Blast.read())
Blast.close()
```

```python
results_Blast= open('blast_DNA_IRS1_file.xml')
blast_records = NCBIXML.parse(results_Blast)
for x in blast_records:
    print(x.database)
    print(x.matrix)
    print(x.gap_penalties)
    print(len(x.alignments))
    print(x.alignments[0])  
```

```python
results_Blast= open('blast_DNA_IRS1_file.xml')
blast_records = NCBIXML.read(results_Blast)
for alignment in  blast_records.alignments:
    for hsp in alignment.hsps:
            print('Alignment')
            print('sequence: ', alignment.title)
            #print('lenght:', alignment.lenght)
            print(hsp.query[0:75] + '...')
            print(hsp.match[0:75] + '...')
            print(hsp.sbjct[0:75] + '...')
            print()
```

```python
E_VALUE_THRESH = 0.001
results_Blast= open('blast_DNA_IRS1_file.xml')
blast_records = NCBIXML.read(results_Blast)
for alignment in  blast_records.alignments:
    #print(alignment.accession)
    for hsp in alignment.hsps:
        if hsp.expect < E_VALUE_THRESH:
            print('****Alignment****')
            print('sequence: ', alignment.title)
            print('lenght:', alignment.length)
            print(hsp.query[0:75] + '...')
            print(hsp.match[0:75] + '...')
            print(hsp.sbjct[0:75] + '...')
            print()
```

```python
import re

results_Blast= open('blast_DNA_IRS1_file.xml')
blast_records = NCBIXML.read(results_Blast)
E_VALUE_THRESH = 0.001
count_preditc, count_homo=0,0
list_filtered_alignments=[]
for alignment in  blast_records.alignments:
    for hsp in alignment.hsps:
        if hsp.expect < E_VALUE_THRESH:
            exist = re.search(r'PREDICTED:\s', alignment.title)
            if exist:
                # print( existe[0] )
                pre = re.match(r'PREDICTED:\s', exist[0] )
                if pre:
                    count_preditc+=1
                    #print(id)
            else:
                list_filtered_alignments.append(alignment.accession)
                homo=re.search(r'Homo\ssapiens',alignment.title)
                if homo:
                    count_homo+=1
print(list_filtered_alignments)
print('In {} alignments, total {} PREDICTED seqs found and remaining {} ids are from Homo sapiens'.format(len(blast_records.alignments),count_preditc, count_homo))
#contagem de "Homo sapiens" não é conclusiva, é meramente informativa. Isto porque, os títulos não seguem sempre o memso padrão.
```

```python
with open('CDS_nucleotide_result_blast.txt', 'w') as f:
    for line in list_filtered_alignments:
        f.write(f"{line}\n")
```
